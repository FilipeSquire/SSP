# Data Enrichment (Coordinates)

```python
from geopy import Nominatim
from geopy.extra.rate_limiter import RateLimiter

geolocator = Nominatim(user_agent='felipe.sql@outlook.com')

# 1- Creating a new column for the city name and then a new dataframe where we have latitude missing data from the original dataset
ax = df[df.LATITUDE.isna() == True][['BAIRRO','CIDADE','UF']].copy()
ax = ax.drop_duplicates() #drop similar address

ax['address'] = ax.BAIRRO+', '+ ax.CIDADE +', '+ ax.UF + ', Brazil'
ax['coordinates'] = ''
```

```python
# 1 - convenient function to delay between geocoding calls
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
# 2- - create location column
ax['coordinates'] = ax['address'].apply(geocode)
# 3 - splitting the info
ax['point'] = ax['coordinates'].apply(lambda loc: tuple(loc.point) if loc else None)
# 4 - split point column into latitude, longitude and altitude columns
ax[['latitude', 'longitude', 'altitude']] = pd.DataFrame(ax['point'].tolist(), index=ax.index)
# 5 - there are some rows where the api failed into fiding the coordinates. 
ax2 = ax[ax.coordinates.isna() == True].copy()
ax2['address'] = ax2.address.str.replace('SP','Sao Paulo')
ax2['coordinates'] = ax2['address'].apply(geocode)
ax2['point'] = ax2['coordinates'].apply(lambda loc: tuple(loc.point) if loc else None)
ax2 = ax2[ax2.coordinates.isna() == False] #excluding the lines where the api failed again to find the coordinates
ax2[['latitude', 'longitude', 'altitude']] = pd.DataFrame(ax2['point'].tolist(), index=ax2.index)
ax[ax.index.isin(ax2.index)][['coordinates','point','latitude','longitude','altitude']] = ax2[['coordinates','point','latitude','longitude','altitude']]
```

```python
print(f'Non identified coordinates {sum(ax.coordinates.isna())/ax.shape[0]*100}')
```

```python
#Joining the datasets
df['address'] = df.BAIRRO+', '+ df.CIDADE +', '+ df.UF + ', Brazil'
df = df.join(ax[['address','latitude','longitude']].set_index('address'), on='address', how='left')
#Creating a copy to work with the dataframe and replace the null values in the original latitude and longitude columns
df2 = df.copy()
mask = df2[df2.LATITUDE.isna() == True]
df2['LATITUDE'] = df2['LATITUDE'].where(df2.LATITUDE.isna() == False, df2.latitude)
df2['LONGITUDE'] = df2['LONGITUDE'].where(df2.LONGITUDE.isna() == False, df2.longitude)
df2[df2.index.isin(mask.index)]
```

```python
#Checking our improvement
print(f'Null latitude share of old dataset: {sum(df.LATITUDE.isna())/df.shape[0]*100}')
print(f'Null latitude share of new dataset: {sum(df2.LATITUDE.isna())/df2.shape[0]*100}')
df = df2
df = df.drop(columns=['latitude','longitude'])
```
